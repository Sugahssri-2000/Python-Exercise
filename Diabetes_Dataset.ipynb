```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("diabetes_dataset.csv")
```

```python
#Load rows and columns of the dataset

print("Head of the dataset:")
print(data.head())

print("Tail of the dataset:")
print(data.tail())
```

# Question 1: What is the distribution of diabetes cases across different age groups?

```python
data['Age_Group'] = pd.cut(data['Age'],
                          bins=[0, 20, 40, 60, 80, 100],
                          labels=['0-20', '21-40', '41-60', '61-80', '81+'])

plt.figure(figsize=(8,6))
sns.countplot(data=data, x='Age_Group', hue='Diabetes_012', palette='Set2')
plt.title('Distribution of Diabetes Cases Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Diabetes Status', labels=['No Diabetes', 'Prediabetes', 'Diabetes'])
plt.show()
```

# Question 2: How does the BMI vary between individuals with and without diabetes?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Diabetes_012', y='BMI', palette='coolwarm')
plt.title('BMI Distribution by Dibetes Status')
plt.xlabel('Daibetes Status')
plt.ylabel('BMI')
plt.xticks(ticks=[0, 1, 2], labels=['No Diabetes', 'Prediabetes', 'Diabetes'])
plt.show()
```
